```python
import pandas as pd
import numpy as np
```

```python
mapping = pd.read_excel(r"Sample Input.xlsx", nrows=19, usecols=range(14))
mapping
```

```python
dtypes = {}

for idx, row in mapping.iterrows():
    typ = row.DataType
    if typ == 'REAL : Float': typ = 'float32'
    elif typ == 'TEXT : Text Data' : typ = 'object'
    elif typ == 'BOOL : Bit Data' : typ = 'bool'
    elif typ == 'INT : Signed 2 Byte Integer' : typ = 'int32'

    dtypes[row.ItemName] = typ
```

```python
data = pd.read_excel(r"Sample Input.xlsx", skiprows=20, dtype=dtypes)
```

```python
tdelta = pd.to_timedelta(data["Milli Sec"], unit="ms")
```

```python
data['event_timestamp'] = data['Date'] + pd.to_timedelta(data['Time'].astype(str)) + tdelta
```

```python
data.drop(['Date', 'Time', 'Milli Sec'], axis = 1, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.set_index('event_timestamp').resample('10S').max()
```

```python
data.reset_index(level=0, inplace=True)
data
```

```python
data[['L_R1_ODG_AUX_B07', 'L_R1_CNT_PASS0']] = data[['L_R1_ODG_AUX_B07', 'L_R1_CNT_PASS0']].astype('int')
```

```python
data = data.melt(id_vars='event_timestamp', var_name='tag__name', value_name='tag__value')
```

```python
data = data.merge(mapping[['ItemId', 'ItemName', 'Comment']], how='inner', left_on='tag__name', right_on='ItemName').drop(['ItemName'], axis=1)
```

```python
def extract_unit(desc):
    ls = desc.split('(')
    try: return ls[1][:-1]
    except: return ''
```

```python
data['tag__unit'] = data['Comment'].apply(extract_unit)
```

```python
data['event_timestamp'] = data['event_timestamp'].dt.strftime('%d-%m-%Y %H:%M')
```

```python
data
```

```python
data.rename(columns = {'ItemId':'tag__id', 'Comment':'tag__desc'}, inplace = True)
```

```python
data = data[['event_timestamp', 'tag__id', 'tag__name', 'tag__desc', 'tag__value', 'tag__unit']]
data = data.set_index('event_timestamp')
data
```

```python
data.to_csv('Output.csv')
```
